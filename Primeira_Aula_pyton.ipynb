```python
# Neste codigo podemos observar que usamos a função Int(), pois sem a mesma o
# Py percebe o codigo como String
Nota_1 = int(input("Digite a primeira nota: "))
Nota_2 = int(input("Digite a segunda nota: "))
Nota_3 = int(input("digite a terceira nota: "))
Nota_4 = int(input("Digite a quarta nota: "))
```

```python
#usando a condição de aprovação para os alunos
media = (Nota_1+Nota_2+Nota_3+Nota_4)/4

if media >= 8.75:
  situacao = "Aprovado, és um Gênio"
else:
  situacao = "Chumbou, que burro!"
```

```python
#dadas as notas, mostramos a média final e a situação do aluno.

print(f"Resultado medio de notas: {media}")

print({situacao})
```
